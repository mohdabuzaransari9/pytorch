```python
import torch
from torch import nn
class Linearregression(nn.Module):
  def __init__(self):
    super().__init__()
    self.weight=nn.Parameter(torch.rand(1,
                                         dtype=torch.float32,
                                         requires_grad=True))
    self.bias=nn.Parameter(torch.rand(1,
                                      dtype=torch.float32,
                                      requires_grad=True))


  def forward(self, x:torch.Tensor)->torch.Tensor:
    return self.weight*x+self.bias
```
